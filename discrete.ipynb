## 1. Найдите наибольший общий делитель чисел а и b

```python
def nod_2(a, b):
    while b != 0:
        a_old = a
        a = b
        b = a_old % b

    return a

a = 50
b = 84

print(f"НОД({a}, {b}) = {nod_2(a, b)}")
```

## 3. Найдите наименьшее общее кратное чисел а и b

```python
def nok_2(a, b):
    # Условие что НОК = 0
    if a == 0 or b == 0:
        result = 0
    else:
        result = a * b // nod_2(a, b) # формула

    return result

a = 46
b = 105

d = nok_2(a, b)

print(f"НОК({a}, {b}) = {d}")
```

## 5. Найдите все простые числа в диапазоне от а до b включительно

```python
def erat(a, b):
    if b < 2 or a > b:
        return []

    start = max(2, a)

    # Список
    prime = [True] * (b + 1) # Все True
    prime[0] = prime[1] = False # первые два значения False


    for i in range(2, int(b ** 0.5) + 1):
        if prime[i]:
            for j in range(i * i, b + 1, i):
                prime[j] = False

    # Список простых чисел от start до b ис обираем все простые
    primes = [i for i in range(start, b + 1) if prime[i]]
    return primes


a = 127
b = 139

result = erat(a, b)
print(f"Простые числа от {a} до {b}, всего {len(result)}")
print(result)
```

## 6. Найдите г - результат а/b в поле вычетов по модулю n, т.е. a/b=r, b*r mod n = a, b!=0

```python
def up_nod(a, b):
    # Базовый случай при a = 0, то НОД = b, а x = 0 и y = 1
    if a == 0:
        nod = b
        x = 0
        y = 1
        return nod, x, y
    
    nod, x1, y1 = up_nod(b % a, a)
    
    # Коэффы
    x = y1 - (b // a) * x1
    y = x1
    
    return nod, x, y


def del_mod(a, b, n):
    if b % n == 0:
        return "деление на ноль"
    
    # НОД и коэффициенты
    nod, x, y = up_nod(b, n)
    
    # Существует обратный?
    if nod != 1:
        return f"деление невозможно, НОД = {nod}"
    
    b_inverse = x % n
    
    result = (a * b_inverse) % n
    
    if (b * result) % n != a % n:
        return "ошибка"
    
    return result


a = 46
b = 82
n = 131

r = del_mod(a, b, n)

print("результат:", r)
```

## 7. Выполните факторизацию n

```python
def sqrt(n):
    if n < 0:
        return False
    num = int(n ** 0.5)
    return num * num == n

def prime(n):
    if n <= 1: 
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

def fermat_factor(n):
    if n == 1:
        return []  # нет множетелей Ферма
    
    if n % 2 == 0:
        return [2] + fermat_factor(n // 2)
    
    if prime(n):
        return [n]
    
    def find_factors(m):
        x = int(m ** 0.5) + 1
        while True:
            diff = x * x - m
            if sqrt(diff):
                y = int(diff ** 0.5)
                return x - y, x + y
            x += 1
    
    a, b = find_factors(n)
    
	# предотвращение беск рекурсии
    factors = []
    if a == 1:
        factors.append(b)  # при a = 1, то b(простое)
    elif b == 1:
        factors.append(a)  # при b = 1, то a(простое)
    else:
        factors = fermat_factor(a) + fermat_factor(b)
    
    return factors


n = 774
factors = fermat_factor(n)
print(f"Факторизация {n}:")
print(f"{n} = {factors}")
```

## 8. Найдите первые 3 числа генерируемые линейным конгруентным генератором(формула х = (a*x+с) % m) с параметрами m=6075, a=106, с=1283, х0

```python
def nok_gener(m, a, c, x_0, N):
    #Генератор чисел по формуле x = (a*x + c) % m
    x = (a * x_0 + c) % m
    k = 0
    while k < N:
        yield x
        x = (a * x + c) % m
        k += 1

m = 6075
a = 106
c = 1283
x_0 = 47
N = 3

gen = nok_gener(m, a, c, x_0, N)

lst = []
for i in range(N):
    lst.append(next(gen))

print(f"Первые {N} числа:")
for i in range(len(lst)):
    print(f"x{i+1} = {lst[i]}")
```

## 9. Зашифруйте число m с помощью алгоритма RSA, если (5 , 119) - открытый ключ

```python
def my_fast_pow(a, ex, mod):
    result = 1
    a = a % mod
    
    while ex > 0:

        if ex % 2 != 0:  # Проверка на нечётность
            result = (result * a) % mod
        
        # Сдвиг и возведение основания в квадрат
        ex = ex // 2
        # Основание в квадрате по модулю
        a = (a * a) % mod
    
    return result

e = 5
n = 119

m = 37

coder = my_fast_pow(m, e, n)
print(f"Зашифрованное сообщение: c = {coder}")
print(f"Проверка со встроенной функцией pow({m}, {e}, {n}): {pow(m, e, n)}")
```
